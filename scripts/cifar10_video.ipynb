# Video of spiking CNN for CIFAR-10

```python
% matplotlib inline
import os
import moviepy.editor as mpy
import matplotlib.pyplot as plt
from scripts.utils import show_architecture, show_labels, show_animated_raster_plots
from scripts.utils import show_input_image, apply_title, load_data_for_video
```

```python
num_samples = 50
num_layers = 8
clamp_duration = 0
simulation_duration = 100
classification_duration = simulation_duration - clamp_duration
video_duration = classification_duration * num_samples
real_classification_time_per_sample_sec = 10
speedup_factor = real_classification_time_per_sample_sec / (classification_duration / 1000) 

path = '/home/rbodo/.snntoolbox/data/cifar10/87.86/log/gui/22'

class_idx_dict = {'0': ['', 'airplane'], '1': ['', 'automobile'], 
                  '2': ['', 'bird'], '3': ['', 'cat'], '4': ['', 'deer'], 
                  '5': ['', 'dog'], '6': ['', 'frog'], '7': ['', 'horse'],
                  '8': ['', 'ship'], '9': ['', 'truck']}
```

```python
def undo_preprocessing(x):
    import numpy as np
    return np.swapaxes(x, 1, 2) * 255.
```

```python
input_images_t, spiketrains_n, top5_labels_t, top5_spikecounts_t, true_labels_t = \
    load_data_for_video(path, classification_duration, simulation_duration, 
                        video_duration, num_samples, num_layers, class_idx_dict,
                        undo_preprocessing)
```

```python
clip_architecture = show_architecture(video_duration / 1000, path)
clip_input = show_input_image(input_images_t, classification_duration / 1000, video_duration / 1000, path)
clip_raster = show_animated_raster_plots(spiketrains_n, classification_duration / 1000, video_duration / 1000, path)
clip_labels = show_labels(top5_labels_t, top5_spikecounts_t, true_labels_t, video_duration / 1000, path)
```

```python
clip_raster = mpy.VideoFileClip(os.path.join(path, 'raster.mp4'))
clip_labels = mpy.VideoFileClip(os.path.join(path, 'labels.mp4'))
clip_input = mpy.VideoFileClip(os.path.join(path, 'images.mp4'))
clip_architecture = mpy.VideoFileClip(os.path.join(path, 'architecture.mp4'))
```

```python
clip_raster_new = clip_raster.margin(10)
clip_labels_new = clip_labels.resize(height=clip_raster.h).margin(10)
clip_input_new = clip_input.resize(width=clip_labels.w).margin(10)
clip_architecture_new = clip_architecture.resize(width=clip_raster.w).margin(10)
```

```python
animation = mpy.clips_array([[clip_architecture_new, clip_input_new],
                             [clip_raster_new, clip_labels_new]],
                            bg_color=(1, 1, 1))
animation.write_videofile(os.path.join(path, 'inference_movie.mp4'), fps=1/1e-3)
```
